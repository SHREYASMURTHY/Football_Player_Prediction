```python
import whisper

def mp3_to_text(mp3_path, model_size="base"):
    # load whisper model
    model = whisper.load_model(model_size)

    print("📝 Transcribing...")
    result = model.transcribe(mp3_path)

    transcript = result["text"]
    print("✅ Transcript (first 500 chars):")
    print(transcript[:500])

    return transcript

# Example usage
text = mp3_to_text("/Users/shreyas/Desktop/NLP/audio/i1vRsbvXxDw.mp3")
```

```python
import re
players_vs_burnley = [
    "Bayindir",
    "Yoro",
    "de Ligt",
    "Shaw",
    "ADiallo",
    "Casemiro",
    "B Fernandes",
    "D Dalot",
    "B Mbeumo",
    "M Mount",
    "M Cunha",
    "N Mazraoui",
    "J Zirkzee",
    "B Sesko",
    "K Mainoo"
]

name_corrections = {
    r"\bBayindir\b": " Bayindir",
    r"\bEuro\b": "Yoro",
    r"\bDelic\b": "De Ligt",
    r"\bLuke Shaw\b": "Shaw",
    r"\bAhmad\b": "Diallo",
    r"\bAmad\b": "Diallo",
    r"\bKazemiro\b": "Casemiro",
    r"\bBruno\b": "Fernandes",
    r"\bDelo[w]?\b": "Dalot",
    r"\bEmbermo\b": "Mbeumo",
    r"\bMount\b": "Mount",
    r"\bMason Mout\b": "Mount",
    r"\bCunya\b": "Cunha",
    r"\bMasrawee\b": "Mazraoui",
    r"\bMaynu\b": "Mainoo",
    r"\bManu\b": "Mainoo",
    r"\bZerkec?[iy]\b": "Zirkzee",
    r"\bSir\s?XC\b": "Zirkzee",
    r"\bSes?h?s?ko\b": "Sesko"
}

for wrong_name, correct_name in name_corrections.items():
    text = re.sub(wrong_name, correct_name, text, flags=re.IGNORECASE)
text  
```

```python
nlp = spacy.load("en_core_web_sm")
doc=nlp(text)
sentences=[]
for sent in doc.sents:
    filtered=[token.text.lower() for token in sent if not token.is_stop and not token.is_punct]
    if filtered:
        sentences.append(" ".join(filtered))
sentences

        
```

```python
from collections import defaultdict

player_dict = defaultdict(list)

for sent in doc.sents:
    sentence_text = sent.text.strip()
    for player in players_vs_burnley:
        if re.search(rf"\b{re.escape(player)}\b", sentence_text, flags=re.IGNORECASE):
            player_dict[player].append(sentence_text)
player_dict.keys()
```

```python
for key,value in player_dict.items():
    print(f"{key}:{value}")
```

```python
import pandas as pd
df = pd.DataFrame({
    "Player": list(player_dict.keys()),
    "Sentences": ["".join(sentences) for sentences in player_dict.values()]
})
df
```

```python
def preprocess(text):
    doc = nlp(text)
    processed_sentences = []

    for sent in doc.sents:
        tokens = []
        for tok in sent: 
            if not tok.is_stop and not tok.is_punct:
                tokens.append(tok.lemma_.lower().strip())

        if tokens:
            processed_sentences.append(" ".join(tokens))

    return processed_sentences
```

```python
df["Processed_Sentences"] = df["Sentences"].apply(lambda x: " | ".join(preprocess(x)))
df
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def get_rating(text):
    score = analyzer.polarity_scores(text)["compound"]  

    rating = round((score + 1) * 5, 1)
    return rating

df["Rating"] = df["Processed_Sentences"].apply(get_rating)

print(df[["Player", "Rating"]])
```
